### 2. Pasangan dengan jarak terjauh (longest shortest-path)

Jadi nilai jarak ini menunjukkan tingkat keterhubungan global jaringan (network connectivity).
Jarak 5 berarti jaringan relatif terpadu, tidak ada individu yang benar-benar terisolasi, meskipun beberapa berada di tepi jaringan sosial.
